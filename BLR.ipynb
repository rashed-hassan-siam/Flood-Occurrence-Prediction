```python
import pandas as pd
from matplotlib import pyplot as plt
from google.colab import files
 
 
uploaded = files.upload()
```

```python
import io
import numpy as np
df= pd.read_csv(io.BytesIO(uploaded['FloodPrediction.csv']))
df['Flood']=df['Flood'].fillna(0)
df.head()
```

```python
plt.scatter(df.Rainfall,df.Flood,marker='+',color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test = train_test_split(df[['Rainfall']],df.Flood,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
