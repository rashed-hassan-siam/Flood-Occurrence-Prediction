```python
from google.colab import files
 
 
uploaded = files.upload()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import io
import numpy as np
df= pd.read_csv(io.BytesIO(uploaded['FloodPrediction.csv']))
df['Flood']=df['Flood'].fillna(0)
df.head()
```

```python
df0=df[df.Flood==0]
df1=df[df.Flood==1]
```

```python
plt.xlabel('Rainfall')
plt.ylabel('Flood')
plt.scatter(df0['Rainfall'],df0['Flood'],color='green',marker='+')
plt.scatter(df1['Rainfall'],df1['Flood'],color='red',marker='+')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(df[['Rainfall']],df.Flood,test_size=0.2)
```

```python
from sklearn.svm import SVC
model= SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
